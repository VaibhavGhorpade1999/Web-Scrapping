```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
base_url="https://www.bikewale.com/used/bikes-in-india/#pn="
Name=[]
Info=[]
Price=[]
for i in range(1,300):
    url=base_url+str(i)
    response=requests.get(url)
    print("scrapping info from page"+str(i))
    soup=BeautifulSoup(response.content,"html.parser")
    name=soup.find_all('a',attrs={"class":"text-truncate text-black"})
    for a in name :
        Name.append(a.text)  
    price=soup.find_all('span',attrs={"class":"font22 text-bold"})
    for b in price:
        Price.append(b.text)
    info=soup.find_all('span',attrs={"class":["model-details-label"]})
    for c in info:
        Info.append(c.text)
Year=[]
for i in Info:
    a=range(0,len(Info)+1,4)
    if Info.index(i) in a:
        Year.append(i)
Kms_driven=[]
for i in Info:
    a=range(1,len(Info)+1,4)
    if Info.index(i) in a:
        Kms_driven.append(i)
Owner_type=[]
for i in Info:
    a=range(2,len(Info)+1,4)
    if Info.index(i) in a:
        Owner_type.append(i)
City=[]
for i in Info:
    a=range(3,len(Info)+1,4)
    if Info.index(i) in a:
        City.append(i)
df=pd.DataFrame({"Name":Name,"Price":Price,"Year":Year,"Kms_Driven":Kms_driven,"Owner_type":Owner_type,"City":City})
df
```
